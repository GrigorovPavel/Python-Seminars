## Задача №61. Решение в группах
1. Определить какое минимальное и максимальное значение median_house_value
2. Показать максимальное median_house_value, где
median_income = 3.1250
3. Узнать какая максимальная population в зоне
минимального значения median_house_value


```python
# 1. Определить какое минимальное и максимальное значение median_house_value
df['median_house_value'].min(), df['median_house_value'].max()
# выводим минимальное и максимальное значения с помощю функций min() и max()
```

```python
# 2. Показать максимальное median_house_value, где median_income = 3.1250
df[df['median_income'] == 3.1250]['median_house_value'].max()
# обращаемся строкам где колонка ['median_income'] == 3.1250 и
# находим максимальное значение ['median_house_value'].max()
```

```python
# 3. Узнать какая максимальная population в зоне минимального значения median_house_value
df[df['median_house_value'] == df['median_house_value'].min()]['population'].max()
# обращаемся к колонке df['median_house_value'] и находим строки с минимальными значениями == df['median_house_value'].min()
# и в этих строках обращаемся к колонке []'population']  и находим максимальное значение ['population'].max()
```

## Task_1_HW
Дан файл california_housing_train.csv.
Определить среднюю стоимость дома , где количество людей от 0 до 500 (population) и сохранить ее в переменную avg.
Используйте модуль pandas.

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv') # В ДЗ указать директорию 'california_housing_train.csv'
df[(df['population'] >0) & (df['population'] <= 500)]['median_house_value'].mean() # mean() - среднее значене
```

```python
avg = df[df['population'] <= 500]['median_house_value'].mean()
```

## Task_2_HW
Дан файл california_housing_train.csv.
Найти максимальное значение переменной "households" в зоне минимального значения переменной min_population и сохраните это значение в переменную max_households_in_min_population.
Используйте модуль pandas.

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv') # В ДЗ указать директорию 'california_housing_train.csv'
min_population = df[df['population'] == df['population'].min()]
max_households_in_min_population = min_population['households'].max()
# max_households_in_min_population
```
