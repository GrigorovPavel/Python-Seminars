## Задача №67. Решение в группах
1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)


```python
penguins.loc[penguins["bill_length_mm"] >= 42, 'height_group'] = 'high'
penguins.loc[(penguins["bill_length_mm"] < 42) & (penguins["bill_length_mm"] > 35), 'height_group']  = 'middle'
penguins.loc[penguins["bill_length_mm"] <= 35, 'height_group'] = 'low'

sns.histplot(data=penguins,x = 'height_group', hue = 'sex')
# В таком варианте колонки какладываются друг на друга
```

```python
sns.histplot(data=penguins,x = 'height_group', hue = 'sex', multiple = 'stack')
# В таком варианте колонки выводятся полностью и разделяются на заданные данные 'hue'
```

```python
sns.histplot(data=penguins,x = 'bill_length_mm', hue = 'height_group', multiple = 'stack')
# количество пингвинов (count) с длиной клюва (bill_length_mm)
```
